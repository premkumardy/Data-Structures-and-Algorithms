0/1 Knapsack
The idea
Suppose we are given a bag/knapsack with a fixed capacity, along with some items' weights and profits we reap by choosing to put that item in the bag. We want to maximize the profit while also ensuring that our backpack doesn't run out of space. The reason this algorithm is called 0/1 is because at each point, we can either choose to include an item or not include it - a binary decision.

We will implemeting a backtracking approach here and will be decreasing capacity as we make further calls.
When capacity is no more or if indx reaches limit, we stop and backtrack(return).

We will be having 2 decicitions on each call, 
1) either to include ith index and reduce i th indx's weight from capacity.
2) not to include i th indx's element and not touch the current capacity(as we did not include the element).

```python
def findMaxprofit(profit,weight,capacity):


    def dfs(i,currProfit,capacity):
        if i >= len(profit):
            return 0
        if capacity-weight[i] < 0:
            return 0
        
        #not including ith element
        print(f"including {i}th profit={profit[i]} weight = {weight[i]}",i+1,currProfit,capacity)
        maxprofit = dfs(i+1,currProfit,capacity)
        
        #including ith element 
        if not capacity-weight[i] < 0:
            maxprofit = max(maxprofit, profit[i] + dfs(i+1,currProfit+profit[i],capacity-weight[i]))
            print(f" not including {i}th profit={profit[i]} weight = {weight[i]}",i+1,currProfit+profit[i],capacity-weight[i])

        return maxprofit

    
    return dfs(0,0,capacity)

profit = [4,4,7,1]
weight = [5,2,3,1]
capacity = 8
findMaxprofit(profit,weight,capacity)
```

```python
def findMaxprofit(profit,weight,capacity):

    cache = [[-1]*(capacity+1) for _ in profit]
    def dfs(i,currProfit,capacity):
        if i >= len(profit):
            return 0
        if cache[i][capacity] != -1:
            return cache[i][capacity]
        
        
        #not including ith element
        # print(f"including {i}th profit={profit[i]} weight = {weight[i]}",i+1,currProfit,capacity)
        cache[i][capacity] = dfs(i+1,currProfit,capacity)
        
        #including ith element 
        if not capacity-weight[i] < 0:
            cache[i][capacity-weight[i]] = profit[i] + dfs(i+1,currProfit+profit[i],capacity-weight[i])
            
        return max(cache[i][capacity],cache[i][capacity-weight[i]])

    
    return dfs(0,0,capacity)

profit = [4,4,7,1]
weight = [5,2,3,1]
capacity = 8
findMaxprofit(profit,weight,capacity)
```

474. Ones and Zeroes

You are given an array of binary strings strs and two integers m and n.

Return the size of the largest subset of strs such that there are at most m 0's and n 1's in the subset.

A set x is a subset of a set y if all elements of x are also elements of y.

 

Example 1:

Input: strs = ["10","0001","111001","1","0"], m = 5, n = 3
Output: 4
Explanation: The largest subset with at most 5 0's and 3 1's is {"10", "0001", "1", "0"}, so the answer is 4.
Other valid but smaller subsets include {"0001", "1"} and {"10", "1", "0"}.
{"111001"} is an invalid subset because it contains 4 1's, greater than the maximum of 3.
Example 2:

Input: strs = ["10","0","1"], m = 1, n = 1
Output: 2
Explanation: The largest subset is {"0", "1"}, so the answer is 2.
 

Constraints:

1 <= strs.length <= 600
1 <= strs[i].length <= 100
strs[i] consists only of digits '0' and '1'.
1 <= m, n <= 100

```python
#works, but time limit exceeded
class Solution(object):
    def findMaxForm(self, strs, m, n):
        """
        here for each ith index, we will choose no to include and choose to include 
        and as we include we will reduce the number of 0's capacity left and number of 1's left and make recursive calls
        if addition of ith index is possible we increment 1 with recursion call, i.e addition of ith element  is possible + return of i+1 th elemt

        we will be comparing return of without including current element and withadding current element, will greedily return the possiblity with
        maximum possible element inclusion while with 1's and 0's capacity given

        """
        OneCount = []
        ZeroCount = []

        for ch in strs:
            OneCount.append(ch.count('1')) #calculating 1's in ith string
            ZeroCount.append(ch.count('0')) #calculating 0's in ith string

        def helpr(i,ones_left,zeros_left):
            if i >= len(strs):
                return 0
            s = helpr(i+1,ones_left,zeros_left)
            if ones_left-OneCount[i] >= 0 and zeros_left-ZeroCount[i] >= 0:
                s = max(s,1+helpr(i+1,ones_left-OneCount[i],zeros_left-ZeroCount[i]))
            return s
        return helpr(0,n,m)

strs = ["10","0001","111001","1","0"]
m = 5
n = 3
Solution().findMaxForm(strs,m,n)
```

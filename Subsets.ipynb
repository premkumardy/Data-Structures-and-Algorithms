Subsets
Solved 
Given an array nums of unique integers, return all possible subsets of nums.

The solution set must not contain duplicate subsets. You may return the solution in any order.

Example 1:

Input: nums = [1,2,3]

Output: [[],[1],[2],[1,2],[3],[1,3],[2,3],[1,2,3]]
Example 2:

Input: nums = [7]

Output: [[],[7]]
Constraints:

1 <= nums.length <= 10
-10 <= nums[i] <= 10

```python
class Solution:
    def subsets(self, nums):
        ans = []
        tmp = []
        n = 0
        def dfs(n):

            if n >= len(nums):
                ans.append(tmp.copy())
                return

            tmp.append(nums[n])
            dfs(n+1)

            tmp.pop()
            dfs(n+1)
        dfs(0)
        return ans
```

Subsets II
Solved 
You are given an array nums of integers, which may contain duplicates. Return all possible subsets.

The solution must not contain duplicate subsets. You may return the solution in any order.

Example 1:

Input: nums = [1,2,1]

Output: [[],[1],[1,2],[1,1],[1,2,1],[2]]
Example 2:

Input: nums = [7,7]

Output: [[],[7], [7,7]]


```python
class Solution:
    def subsetsWithDup(self, nums):
        nums.sort()
        ans,tmp = [],[]

        def dfs(n):
            if n >= len(nums):
                ans.append(tmp.copy())
                return
            
            tmp.append(nums[n])
            dfs(n+1)
            rmoved = tmp.pop()

            while n < len(nums)-1 and nums[n+1] == rmoved:
                n+=1

            dfs(n+1)
        dfs(0)
        return ans
```

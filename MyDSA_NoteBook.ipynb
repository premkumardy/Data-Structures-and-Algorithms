```python
nums = [-1,0,3,5,9,12]

def BinarySearch(nums,target):
    l,r = 0, len(nums)-1

    while l<=r:
        m = (l+r)//2
        print("nums[m] =",nums[m],"m=",m,"l=",l,"r=",r)
        if nums[m] == target:
            return m
        elif target > nums[m]:
            l = m+1
        elif target < nums[m]:
            r = m-1
                  
    return None

print(BinarySearch(nums,5))
```

```python
class ListNode:
    def __init__(self, val):
        self.val = val
        self.next = None

# Implementation for Singly Linked List
class LinkedList:
    def __init__(self):
        # Init the list with a 'dummy' node which makes 
        # removing a node from the beginning of list easier.
        self.head = ListNode(-1)
        self.tail = self.head
    
    def insertEnd(self, val):
        self.tail.next = ListNode(val)
        self.tail = self.tail.next

    def remove(self, index):
        i = 0
        curr = self.head
        while i < index and curr:
            i += 1
            curr = curr.next
        
        # Remove the node ahead of curr
        if curr and curr.next:
            if curr.next == self.tail:
                self.tail = curr
            curr.next = curr.next.next

    def print(self):
        curr = self.head.next
        while curr:
            print(curr.val, " -> ", end="")
            curr = curr.next
        print()

head = LinkedList()
head.print()


```

```python
def isValid(s):

    stack = []
    hm = {')':'(',']':'[','}':'{'}
    for chr in s:
        if chr not in hm:
            stack.append(chr)
        else:
            if stack and stack[-1] == hm[chr]:
                stack.pop()
            else:
                return False
    return not stack
        
isValid("([{}]())[]")
```

```python

count = []

def Fibonacci(n):
    global count  # Ensure count is declared global inside the function
    count.append(n)
    if n <= 1:
        return n

    return Fibonacci(n-1) + Fibonacci(n-2)

print(Fibonacci(5))
print("count =", count)  # Print count after the function call to see the updates
```

3. Longest Substring Without Repeating Characters

Given a string s, find the length of the longest substring without duplicate characters.

 

Example 1:

Input: s = "abcabcbb"
Output: 3
Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.
Example 2:

Input: s = "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.
Example 3:

Input: s = "pwwkew"
Output: 3
Explanation: The answer is "wke", with the length of 3.
Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.
 

Constraints:

0 <= s.length <= 5 * 104
s consists of English letters, digits, symbols and spaces.

```python
"""
The logic is that we will have two pointer l and r,both begning at same point 0

we will iterate through the string
if s[i] does not exist in set we will be adding it to the set and update the ans by checking the distance between l-curr index(i) + 1 

while s[i] exist in set:
    remove element from s[l], remove all the elements left to the previous duplicate element
    l+=1

when we encounter duplicate we will remove first occurence of the duplicate string along with previous strings that were added in the set,
so they wont cause any issue in further checks 
"""
```

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        ans = 0
        l = 0
        Set = set()
        for i in range(len(s)):
            if s[i] in Set:
                l += 1
            else:
                Set.add(s[i])
            if i-l+1 > ans:
                ans = i-l+1
        return ans
    
Solution().lengthOfLongestSubstring("pwwkew")

```

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        l,maxLength, hs = 0,0,set()

        for r in range(len(s)):

            while s[r] in hs:
                hs.remove(s[l])
                l = l+1
            hs.add(s[r])
            maxLength = max(maxLength,r-l+1)
        return maxLength
```

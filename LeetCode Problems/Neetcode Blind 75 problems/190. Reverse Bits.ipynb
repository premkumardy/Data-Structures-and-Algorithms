190. Reverse Bits

Reverse bits of a given 32 bits signed integer.

 

Example 1:

Input: n = 43261596

Output: 964176192

Explanation:

Integer	Binary
43261596	00000010100101000001111010011100
964176192	00111001011110000010100101000000
Example 2:

Input: n = 2147483644

Output: 1073741822

Explanation:

Integer	Binary
2147483644	01111111111111111111111111111100
1073741822	00111111111111111111111111111110
 

Constraints:

0 <= n <= 231 - 2
n is even.

```python
#optimized
class Solution(object):
    def reverseBits(self, n):
        """
        the idea is to push one bit left to make space for new bit
        if the current bit in n's end is 1 then we will add ..0001 to res (which was 0b000 initially)
        if not 1 then we will again push the res to last bit of res to make space for new bit, making the previous bit remain 0
        return res
        """
        
        res = 0
        for i in range(32):
            res = res << 1
            res += (n & 1)
            n = n >> 1
        return res
    
    
Solution().reverseBits(43261596)
```

```python
class Solution(object):
    def reverseBits(self, n):
        """
        :type n: int
        :rtype: int
        """
        res = 0
        for i in range(33):
            offset = 1
            if n & 1:
                #offset << (32-i) is wrong because here offset 1 already has 1 at 32nd position we need to shift till 31st position
                res |= (offset << (31-i)) 
                
            n = n >>1
        return res
    
Solution().reverseBits(43261596)

    
```

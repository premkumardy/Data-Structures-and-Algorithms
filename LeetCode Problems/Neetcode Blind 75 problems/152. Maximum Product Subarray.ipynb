152. Maximum Product Subarray

Given an integer array nums, find a subarray that has the largest product, and return the product.

The test cases are generated so that the answer will fit in a 32-bit integer.

Note that the product of an array with a single element is the value of that element.

 

Example 1:

Input: nums = [2,3,-2,4]
Output: 6
Explanation: [2,3] has the largest product 6.
Example 2:

Input: nums = [-2,0,-1]
Output: 0
Explanation: The result cannot be 2, because [-2,-1] is not a subarray.
 

Constraints:

1 <= nums.length <= 2 * 104
-10 <= nums[i] <= 10
The product of any subarray of nums is guaranteed to fit in a 32-bit integer.

```python
class Solution(object):
    def maxProduct(self, nums):
        """
        the requirement is to return maximum product subbary in a given array
        the issue is that we will be having negative and 0 values in the array
        neglecting negative value is an issue, if consecutive negative value appears it might
        -turnout to be a large product ex:- -1,2,-4,-5,-6

        Here we use a dynamic programming approach
        

        """
        ans = nums[0]
        curmin,curmax = 1,1
        for i in nums:
            tmp = curmax * i
            curmax = max(i,curmax*i,curmin*i)
            curmin = min(i,tmp,curmin*i)
            ans = max(ans,curmax)
        return ans
    
Solution().maxProduct([-3,-1,-1])
```

```python
#this is a valid bruteforce solution, but time complexity is O(N^2)
class Solution(object):
    def maxProduct(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        ans = float("-inf")
        for i in range(len(nums)):
            if nums[i] > ans:
                ans = nums[i]
            tmp = nums[i]
            for j in range(i+1,len(nums)):
                tmp *= nums[j]
                ans = max(ans,tmp)
        return ans
Solution().maxProduct([-3,-1,-1])
```

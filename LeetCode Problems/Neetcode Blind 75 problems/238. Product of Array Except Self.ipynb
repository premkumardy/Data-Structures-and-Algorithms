238. Product of Array Except Self

Given an integer array nums, return an array answer such that answer[i] is equal to the product of all the elements of nums except nums[i].

The product of any prefix or suffix of nums is guaranteed to fit in a 32-bit integer.

You must write an algorithm that runs in O(n) time and without using the division operation.

 

Example 1:

Input: nums = [1,2,3,4]
Output: [24,12,8,6]
Example 2:

Input: nums = [-1,1,0,-3,3]
Output: [0,0,9,0,0]
 

Constraints:

2 <= nums.length <= 105
-30 <= nums[i] <= 30
The input is generated such that answer[i] is guaranteed to fit in a 32-bit integer.
 

Follow up: Can you solve the problem in O(1) extra space complexity? (The output array does not count as extra space for space complexity analysis.)

```python
"""
the requirement is to return an arr where its has product of all elements of nums except current element nums[i]

we can use brute force an use nested loops to find, but here input range is from 10^5

an O(N) solution would be to main tain a left and right array which will have product of elements to left and
right of current element 

when we muliply the left[i] * righ[i] it will be equalent of product of whole nums except nums[i]

"""
```

```python
class Solution(object):
    def productExceptSelf(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        l_arr,r_arr = [0]*len(nums),[0]*len(nums)
        l,r = 1,1
        for i in range(len(nums)):
            j = -i-1
            l_arr[i] = l
            r_arr[j] = r
            l *= nums[i]
            r *= nums[j]

        return [l*r for l,r in zip(l_arr,r_arr)]
```

```python
class Solution(object):
    def productExceptSelf(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        ans = [0]*len(nums)
        l = 1
        r = 1
        l_arr,r_arr = [0]*len(nums),[0]*len(nums)
        for i in range(len(nums)):
            j = -i-1
            l_arr[i] = l
            r_arr[j] = r
            l *= nums[i]
            r *= nums[j]
        for i in range(len(nums)):
            ans[i] = l_arr[i]*r_arr[i]
        return ans

nums =  [-1,1,0,-3,3]
Solution().productExceptSelf(nums)
```

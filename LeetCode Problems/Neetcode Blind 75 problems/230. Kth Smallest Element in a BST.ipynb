230. Kth Smallest Element in a BST

Given the root of a binary search tree, and an integer k, return the kth smallest value (1-indexed) of all the values of the nodes in the tree.

 

Example 1:



Input: root = [3,1,4,null,2], k = 1
Output: 1
Example 2:



Input: root = [5,3,6,2,4,null,null,1], k = 3
Output: 3
 

Constraints:

The number of nodes in the tree is n.
1 <= k <= n <= 104
0 <= Node.val <= 104

```python
"""
The intuition here is to replicate the DFS search in an iterative manner

first we will travesre all the way to left, while doing so we will append the nodes we travese in a stack, so we can pop backup when a leaf is met
once we pop back up from left most leaf, we can start travesing left from the right side of the curren node and reappeat the entire process

while the process is being repeated, we will decrement the k value for each smallest node we pop, when k value is 0, curr node value is returned which will be the kth smallest element
"""

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def kthSmallest(self, root, k):

        stack = []
        curr = root
        while curr or stack:

            while curr:
                stack.append(curr)
                curr = curr.left

            curr = stack.pop()

            k -= 1
            if not k:
                return curr.val
            
            curr = curr.right
```

```python


"""
the intuition here is to do inorder dfs and store the inorder values in an array
we will break the recursion with base condition where if the length of the inorder values appended is greater than the kth smallest value
we will return and end the recursive calls

return the kth smallest element in arr but makeing the k as 0 based indexing

"""
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def kthSmallest(self, root, k):
        arr = []

        def dfs(root):
            if len(arr) >= k:
                return

            if not root:
                return

            dfs(root.left)
            arr.append(root.val)
            dfs(root.right)

        dfs(root)

        return arr[k-1]
```

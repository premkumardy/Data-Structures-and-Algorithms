104. Maximum Depth of Binary Tree

Given the root of a binary tree, return its maximum depth.

A binary tree's maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.

 

Example 1:


Input: root = [3,9,20,null,null,15,7]
Output: 3
Example 2:

Input: root = [1,null,2]
Output: 2
 

Constraints:

The number of nodes in the tree is in the range [0, 104].
-100 <= Node.val <= 100

Requirement:- return the maximum depth or level of Binary tree

Intuition:

method 1:
Recursively check for maximum return value from left and right child and add to current node 's 1 value and return, if a leaf is met, return 0

when a non root is given it will return 0 for base case:
        
        if not root:
            return 0

for each level it return 1 + maxmimum ( recursive root.left call , recursive root.right call )

```python
"""
Intuition:-

The idea is to perform DFS traversal. when iterataing a curr node we will taking the height of the node as 1 and will we returning maximum of the recursive return from root.left and root.right

if a None root is give as input, the traversal will not happen and since its not valid node, it returns zero


"""
#TimeComplexity O(log(n)),O(n) in worst

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right""
from collections import deque
class Solution:
    def maxDepth(self, root) -> int:
        if not root:
            return 0
        return 1 + max(self.maxDepth(root.left),self.maxDepth(root.right))
                

class TreeNode:
    def __init__(self,data=None,left=None,right=None):
        self.data = data
        self.left = left
        self.right = right
#recursive BST insert
import random
def insertBST(root,data):
    if not root:
        return TreeNode(data=data)
    if data < root.data:
        root.left = insertBST(root.left,data)
    elif data > root.data:
        root.right = insertBST(root.right,data)
    return root

root = TreeNode(4)
x = [x for x in range(0,10)]
for i in x:
    insertBST(root,i)

Solution().maxDepth(root)
```

Intuition:

method 1:

if the given root is not valid return 0 since there is no child or level

create a que and append root node

initialize level to 0

till que is not empty repeat:
    increment counter by 1
    iterate only for curren tlength of que:
        iteratively go through each node and check for left and right node and append to que and pop from left

return the level variiable


Intuition of this method is that the while loop iterates for each level in a tree and for loop inside iterates for each node in current level



```python
#TimeComplexity O(log(n)),O(n) in worst

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
from collections import deque
class Solution:
    def maxDepth(self, root) -> int:
        
        if not root:
            return 0
        
        que = deque()
        que.append(root)
        level = 0
        while que:
            for i in range(len(que)):
                curr = que.popleft()
                if curr:
                    if curr.left:
                        que.append(curr.left)
                    if curr.right:
                        que.append(curr.right)
            level += 1
        return level
                

class TreeNode:
    def __init__(self,data=None,left=None,right=None):
        self.data = data
        self.left = left
        self.right = right
#recursive BST insert
import random
def insertBST(root,data):
    if not root:
        return TreeNode(data=data)
    if data < root.data:
        root.left = insertBST(root.left,data)
    elif data > root.data:
        root.right = insertBST(root.right,data)
    return root

root = TreeNode(4)
x = [x for x in range(0,10)]
for i in x:
    insertBST(root,i)

Solution().maxDepth(root)
```

```python
"""
Requirement is to return strings grouped together by anagrams, ie strings that can rearranged to form the other, vice versa

the intuition is to group the stings together in to dictionary's value using char count as unique key

we will be taking list updated with each letter's index as 0 based index by taking a letter and reducing the offset and creating a key,
and appending strings which share the same char couunt key in hm

once all strings has been mapped, we will return list of hash map's value

"""

class Solution:
    def groupAnagrams(self, strs):
        offset = ord('a')
        hm = {}
        for s in strs:
            count = [0]*26
            for ch in s:
                count[ord(ch)-offset]+=1
            count = tuple(count)
            if not count in hm:
                hm[count] = [s]
            else:
                hm[count].append(s)
        ans = []
        for v in hm.values():
            ans.append(v)
        return ans
    
Solution().groupAnagrams(["act","pots","tops","cat","stop","hat"])
#time complexity O(n*m), m =number of chars in each word
```

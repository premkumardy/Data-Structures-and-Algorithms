191. Number of 1 Bits

Given a positive integer n, write a function that returns the number of set bits in its binary representation (also known as the Hamming weight).

 

Example 1:

Input: n = 11

Output: 3

Explanation:

The input binary string 1011 has a total of three set bits.

Example 2:

Input: n = 128

Output: 1

Explanation:

The input binary string 10000000 has a total of one set bit.

Example 3:

Input: n = 2147483645

Output: 30

Explanation:

The input binary string 1111111111111111111111111111101 has a total of thirty set bits.

 

Constraints:

1 <= n <= 231 - 1

```python
#time complexity O(k) k= number of 1 in the that bin of n
class Solution(object):
    def hammingWeight(self, n):
        """
        the grand idea is to run the loop for the number of times the 1 exist in the given n
        we will reduce 1 from n and do n&(n-1), Ex. what this does is it take 
        n = 20
        n&(n-1) = 
        10100 
        10011
        -----
        10000  ->  1 bit gone

        n&(n-1) = 
        10000
        01111
        -----
        00000 -> 1 bit gone



        n =  45
        0b101101
        0b101100
        --------
        0b101100

        0b101100
        0b101011
        --------
        0b101000

        0b101000
        0b100111
        --------
        0b100000

        0b100000
        0b011111
        --------
        0b000000



        """
        c = 0
        while n:
            # print(f"n({n}) = {bin(n)}, n-1({n-1}) = {bin(n-1)}")
            # print(f"n&(n-1) {n&(n-1)} = {bin(n&(n-1))}")
            print(bin(n))
            print(bin(n-1))
            print(bin(n&(n-1)))
            print()
            n = n & (n-1)

            c += 1
        return c
Solution().hammingWeight(45)
```

```python
class Solution(object):
    def hammingWeight(self, n):
        """
        the requirement is to return the count of 1's in the given n's binary value
        we know that if we do n & 1 (1's bin is ..0001) it will be returning 1 if its odd and it returns 1 we know that
        last bit is 1, this can be used as ad vantage for counting the last 1 bit

        >> 1 operator shift bitt to right side by 1 and removes the last bit, this can be used for poping the last vale

        we run this in a loop and when the number is all ..0000 then the loop breaks and we return the count of bits
        """
        c = 0
        while n:
            if n & 1:
                c += 1
            n = n>>1
        return c
Solution().hammingWeight(2147483645)
```

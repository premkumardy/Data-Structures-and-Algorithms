
70. Climbing Stairs

You are climbing a staircase. It takes n steps to reach the top.

Each time you can either climb 1 or 2 steps. In how many distinct ways can you climb to the top?

 

Example 1:

Input: n = 2
Output: 2
Explanation: There are two ways to climb to the top.
1. 1 step + 1 step
2. 2 steps
Example 2:

Input: n = 3
Output: 3
Explanation: There are three ways to climb to the top.
1. 1 step + 1 step + 1 step
2. 1 step + 2 steps
3. 2 steps + 1 step
 

Constraints:

1 <= n <= 45

```python
# space optimized DP : bottom up
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        if n < 3:
            return n
        dp = [1,2]
        i = 3
        while i <=n:
            tmp = dp[1]
            dp[1] = dp[0]+dp[1]
            dp[0] = tmp
            i += 1
        

        # print(c)
        return dp[1]
        
Solution().climbStairs(10)

```

```python
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """
        cache = {0:0,1:1,2:2}
        # c = 0
        def helper(i):
            # nonlocal c
            # c += 1
            if i in cache:
                return cache[i]
            cache[i] = helper(i-1)+helper(i-2)
            return cache[i]

        ans = helper(n)
        # print(c)
        return ans
        
Solution().climbStairs(10)
```

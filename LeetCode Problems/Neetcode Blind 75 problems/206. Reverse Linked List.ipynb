
206. Reverse Linked List

Given the head of a singly linked list, reverse the list, and return the reversed list.

 

Example 1:



Input: head = [1,2,3,4,5]
Output: [5,4,3,2,1]
Example 2:



Input: head = [1,2]
Output: [2,1]
Example 3:

Input: head = []
Output: []
 

Constraints:

The number of nodes in the list is the range [0, 5000].
-5000 <= Node.val <= 5000

```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def reverseList(self, head):
        """
        the intuition here is to iterate through linked list
        we will be maintaing 2 pointers curr and prev

        in the while loop we will be iterating to next node of the list
        while doing so 
        1) store reference to upcoming node in tmp
        2) since we secured the address of upcoming node, we can set curr.next 
        -with prev node's address (for first node of iteration, prev will be None)
        3)store curr node address which is [curr,curr.next =-> prev] to prev
        4)set curr to the upcoming node

        while loop will run till curr becomes None

        """

        prev = None
        curr = head

        while curr:
            tmp = curr.next
            curr.next = prev
            prev = curr
            curr = tmp
            
        return prev
            

```

```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
head = ListNode()
curr = head
for i in range(5+1):

    curr.next = ListNode(i)
    curr = curr.next



root = Solution().reverseList(head.next)

vals = []
while root:
    vals.append(str(root.val))
    root = root.next
print("->".join(vals))
```

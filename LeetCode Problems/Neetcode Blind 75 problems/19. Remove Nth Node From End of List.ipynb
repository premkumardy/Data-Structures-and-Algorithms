Given the head of a linked list, remove the nth node from the end of the list and return its head.

 

Example 1:



Input: head = [1,2,3,4,5], n = 2
Output: [1,2,3,5]
Example 2:

Input: head = [1], n = 1
Output: []
Example 3:

Input: head = [1,2], n = 1
Output: [1]
 

Constraints:

The number of nodes in the list is sz.
1 <= sz <= 30
0 <= Node.val <= 100
1 <= n <= sz


```python
#one pass method #o(n) TC

"""
the intuition of theis method is to travesere R pointer n+1 times and traverse L and R pointer both at same times till R is a valid node
by the time R reaches None, L would we before the node that needs to be removed

the whole above logic is to make l point at the count - nth position and remove the nth node

"""
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeNthFromEnd(self, head, n):
        if not head:
            return 

        dummy = ListNode()
        dummy.next = head
        l = r = dummy

        for i in range(n+1):
            r = r.next
        
        while r:
            l = l.next
            r = r.next
        l.next = l.next.next

        return dummy.next
```

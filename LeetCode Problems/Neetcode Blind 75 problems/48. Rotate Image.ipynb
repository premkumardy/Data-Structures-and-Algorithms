You are given an n x n 2D matrix representing an image, rotate the image by 90 degrees (clockwise).

You have to rotate the image in-place, which means you have to modify the input 2D matrix directly. DO NOT allocate another 2D matrix and do the rotation.

 

Example 1:



Input: matrix = [[1,2,3],[4,5,6],[7,8,9]]
Output: [[7,4,1],[8,5,2],[9,6,3]]
Example 2:



Input: matrix = [[5,1,9,11],[2,4,8,10],[13,3,6,7],[15,14,12,16]]
Output: [[15,13,2,5],[14,3,4,1],[12,6,8,9],[16,7,10,11]]
 

Constraints:

n == matrix.length == matrix[i].length
1 <= n <= 20
-1000 <= matrix[i][j] <= 1000

```python
"""
The intution is to swap values in all 4 corner of the matrix in a rotating circular way for len(matrix)-1 times
we will store the value of top left in a temporary variable and will swap values in below order

            *this operation is done for len(matrix) - 1 times
                tmp = matrix[top][l+i]

                matrix[top][l+i] = matrix[bottom-i][l]

                matrix[bottom-i][l] = matrix[bottom][r-i]

                matrix[bottom][r-i] = matrix[top+i][r]

                matrix[top+i][r] = tmp

            once we complete outer level we want to move in one level inside that is done by 

            while l < r: # the loop runs till inner matrix > 1 
                
                swapping operation, 

                l+=1
                r-=1

the logic is iterative square swapping in layers


"""

class Solution:
    def rotate(self, matrix) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        
        l,r = 0,len(matrix)-1 #0,2
        while l < r:

            for i in range(r-l): # 0,1
                top , bottom = l,r

                tmp = matrix[top][l+i]

                matrix[top][l+i] = matrix[bottom-i][l]

                matrix[bottom-i][l] = matrix[bottom][r-i]

                matrix[bottom][r-i] = matrix[top+i][r]

                matrix[top+i][r] = tmp

            l += 1
            r -= 1
        
        return matrix


matrix = [[5,1,9,11],[2,4,8,10],[13,3,6,7],[15,14,12,16]]
Solution().rotate(matrix)
```

```python
#this is fine, but we want to perform the operation with O(1) space complexity
"""
the intuition here is to turn the square matrix to 90 degree, one left turn 

So this formula, we have turned the matrix by 90 deg
                # 0 0 = 2 0
                # 0 1 = 1 0
                # 0 2 = 0 0

                # 1,0 = 2 1
                # 1,1 = 1 1
                # 1,2 = 0 1

"""
class Solution:
    def rotate(self, matrix) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        
        n = len(matrix)-1

        res = [[0]*(n+1) for i in range(n+1)]

        for i in range(n+1):
            for j in range(n+1):
                res[i][j] = matrix[n-j][i] 
        print(res)

matrix = [[5,1,9,11],[2,4,8,10],[13,3,6,7],[15,14,12,16]]
Solution().rotate(matrix)
```

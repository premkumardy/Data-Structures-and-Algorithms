100. Same Tree

Given the roots of two binary trees p and q, write a function to check if they are the same or not.

Two binary trees are considered the same if they are structurally identical, and the nodes have the same value.

 

Example 1:



Input: p = [1,2,3], q = [1,2,3]
Output: true
Example 2:



Input: p = [1,2], q = [1,null,2]
Output: false
Example 3:



Input: p = [1,2,1], q = [1,1,2]
Output: false
 

Constraints:

The number of nodes in both trees is in the range [0, 100].
-104 <= Node.val <= 104

```python
"""
the requirement is to check if structure and value of given tree node's are same
return True if same else False

the first solution is following dfs approach

    def isSameTree(self, p, q):
        if not p and not q: #if the current p and q both nodes are not valid and None we will return True since bot are same
            return True
        
        if p and q and p.val == q.val: # if p not is valid and q node is valid and value's are same then we will go further
            
            return self.isSameTree(p.left,q.left) and self.isSameTree(p.right,q.right) #into left and then right branch
        else: # if any other case like value is not equal or any one of p or q node is None we will return False because there is a mismatch between P and Q
            return False
        return True

"""

# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def isSameTree(self, p, q):
        if not p and not q:
            return True
        
        if p and q and p.val == q.val:
            return self.isSameTree(p.left,q.left) and self.isSameTree(p.right,q.right)
        else:
            return False
        return True
```

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
from collections import deque
class Solution:
    def isSameTree(self, p, q):
        if not p and not q:
            return True

        que = [[p,q]]
        while que:
            for i in range(len(que)):
                l,r = que.pop()
                if not l and not r:
                    continue
                if l and r and l.val == r.val:
                    que.append((l.left,r.left))
                    que.append((l.right,r.right))
                else:
                    return False
        return True
        
```

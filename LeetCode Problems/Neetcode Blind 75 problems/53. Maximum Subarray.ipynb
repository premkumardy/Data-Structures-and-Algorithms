
53. Maximum Subarray

Given an integer array nums, find the subarray with the largest sum, and return its sum.

 

Example 1:

Input: nums = [-2,1,-3,4,-1,2,1,-5,4]
Output: 6
Explanation: The subarray [4,-1,2,1] has the largest sum 6.
Example 2:

Input: nums = [1]
Output: 1
Explanation: The subarray [1] has the largest sum 1.
Example 3:

Input: nums = [5,4,-1,7,8]
Output: 23
Explanation: The subarray [5,4,-1,7,8] has the largest sum 23.
 

Constraints:

1 <= nums.length <= 105
-104 <= nums[i] <= 104


```python
class Solution(object):
    def maxSubArray(self, nums):
        """
        we need to return the subarray sum with max sum value

        we will iterate through nums, and we will maintain 2 variable curr - representing current sum,
        maxSum representing maximum sum we can have from the sub array

        we will have a problem that some value will be negative and adding them can make it 0 or less than zero
        To handle the above issue we can reset the curr to 0, kinda like making the left pointer to point the next indx nums[indx]
        from which the curr < 1 happened.
        during each iteration we will greedly store the maximum sum in the maxSum and return the maximum sum
        """
        curr = 0
        maxSum = nums[0]
        for i in nums:
            if curr < 0:
                curr = 0
            curr += i
            if curr > maxSum:
                maxSum = curr
        return maxSum
        
nums = [-2,1,-3,4,-1,2,1,-5,4]
Solution().maxSubArray(nums)
```

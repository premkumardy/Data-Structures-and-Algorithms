153. Find Minimum in Rotated Sorted Array

Suppose an array of length n sorted in ascending order is rotated between 1 and n times. For example, the array nums = [0,1,2,4,5,6,7] might become:

[4,5,6,7,0,1,2] if it was rotated 4 times.
[0,1,2,4,5,6,7] if it was rotated 7 times.
Notice that rotating an array [a[0], a[1], a[2], ..., a[n-1]] 1 time results in the array [a[n-1], a[0], a[1], a[2], ..., a[n-2]].

Given the sorted rotated array nums of unique elements, return the minimum element of this array.

You must write an algorithm that runs in O(log n) time.

 

Example 1:

Input: nums = [3,4,5,1,2]
Output: 1
Explanation: The original array was [1,2,3,4,5] rotated 3 times.
Example 2:

Input: nums = [4,5,6,7,0,1,2]
Output: 0
Explanation: The original array was [0,1,2,4,5,6,7] and it was rotated 4 times.
Example 3:

Input: nums = [11,13,15,17]
Output: 11
Explanation: The original array was [11,13,15,17] and it was rotated 4 times. 
 

Constraints:

n == nums.length
1 <= n <= 5000
-5000 <= nums[i] <= 5000
All the integers of nums are unique.
nums is sorted and rotated between 1 and n times.

```python
class Solution(object):
    def findMin(self, nums):
        """
        Observation: we are given a sorted array where left segment might be rotated
        if left most element is less than right most element, we can say that its not rotated and its sorted
        nums[0] < nums[-1] -> nums is sorted and untouched

        when we take a mid point and if is greater than left it means that we are in the left part of roatated sigment and we can start looking right part,
        but before shifting focus to right segment, what if mid is the smallest ? , use greedy and update res variable on each iteration.
        when looop break, we would have updated res with min value.
        
        """
        l = 0
        r = len(nums)-1
        if nums[0] < nums[r]:
            return nums[0]
        res = nums[0]
        while l <= r:
            if nums[l] < nums[r]:
                res = min(res,nums[l])

            m = (l+r)//2
            if nums[m] < res:
                res = nums[m]
            if nums[m] >= nums[l]: # here the use of >= is crucial, coz in case of [2,1] is input
                                    #then the mid value is (0+1)//2 = 0 and mid = 2 and nums[l] = 2 and we return 2 as ans which is wrong
                l = m+1
            else:
                r = m-1 
        return res
Solution().findMin([4,5,6,7,0,1,2])
```

```python
class Solution(object):
    def findMin(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        l = 0
        r = len(nums)-1
        if nums[0] < nums[r]:
            return nums[0]
        res = nums[0]
        while l <= r:
            if nums[l] < nums[r]:
                res = min(res,nums[l])

            m = (l+r)//2
            if nums[m] < res: # this works too
                res = nums[m]
            if nums[m] < nums[r]:
                r = m-1
            else:
                l = m+1 
        return res
Solution().findMin([2,1])
```

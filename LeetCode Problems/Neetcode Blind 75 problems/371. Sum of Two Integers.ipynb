371. Sum of Two Integers

Given two integers a and b, return the sum of the two integers without using the operators + and -.

 

Example 1:

Input: a = 1, b = 2
Output: 3
Example 2:

Input: a = 2, b = 3
Output: 5
 

Constraints:

-1000 <= a, b <= 1000

```python
"""
In a 32-bit system, you have $2^{32}$ possible combinations ($4,294,967,296$ values). 
However, computers need a way to represent both positive and negative numbers within those same 32 bits.

1. The Sign Bit
To handle negative numbers, computers use the most significant bit (the 31st bit, counting from 0) as a "Sign Bit."

1)If the 31st bit is 0, the number is positive.

2)If the 31st bit is 1, the number is negative.

0xFFFFFFFF = 32 consecutive 1's
0xFFFFFFFF = 0 followed by 31 concecutive ones

any number & 0xFFFFFFFF = same number

...00010
...11111
-------- &
...00010

we do & with the bit mask to ensure it is not overflowing with negative number

"""


a,b = -8,16

# print(bin(1),bin(1))

while b & 0xFFFFFFFF:
    tmp = (a&b) << 1
    a = (a^b) & 0xFFFFFFFF
    b = tmp & 0xFFFFFFFF
    # print(f"a^b = {bin(a)},a&b << 1 = {bin(b)}")

if a <= 0x7FFFFFFF:
    print(a)
else:
    print(~(a^0xFFFFFFFF))


```

```python
class Solution(object):
    def getSum(self, a, b):
        """
        :type a: int
        :type b: int
        :rtype: int
        """
        while b & 0xFFFFFFFF:
            tmp = (a&b) << 1
            a = (a^b) & 0xFFFFFFFF
            b = tmp & 0xFFFFFFFF
        return a if a <= 0x7FFFFFFF else ~(a^0xFFFFFFFF)
    

a,b = -8,16
Solution().getSum(a,b)
```

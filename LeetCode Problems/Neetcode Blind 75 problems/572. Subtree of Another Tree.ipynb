Given the roots of two binary trees root and subRoot, return true if there is a subtree of root with the same structure and node values of subRoot and false otherwise.

A subtree of a binary tree tree is a tree that consists of a node in tree and all of this node's descendants. The tree tree could also be considered as a subtree of itself.

 

Example 1:



Input: root = [3,4,5,1,2], subRoot = [4,1,2]
Output: true
Example 2:



Input: root = [3,4,5,1,2,null,null,null,null,0], subRoot = [4,1,2]
Output: false
 

Constraints:

The number of nodes in the root tree is in the range [1, 2000].
The number of nodes in the subRoot tree is in the range [1, 1000].
-104 <= root.val <= 104
-104 <= subRoot.val <= 104

```python
"""
Refer back to LC100 same Tree

the intuition is to traverse each root with preorder traversal and check if the current root we are at is the begining of subroot,
we will do that using SameTree function, it returns true if the entire tree structure is same

Intuition of SameTree function
the requirement is to check if structure and value of given tree node's are same
return True if same else False

the first solution is following dfs approach

    def isSameTree(self, p, q):
        if not p and not q: #if the current p and q both nodes are not valid and None we will return True since bot are same
            return True
        
        if p and q and p.val == q.val: # if p is valid and q node is valid and value's are same then we will go further
            
            return self.isSameTree(p.left,q.left) and self.isSameTree(p.right,q.right) #into left and then right branch
            
        else: # if any other case like value is not equal or any one of p or q node is None we will return False because there is a mismatch between P and Q
            return False
        return True

"""

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isSubtree(self, root, subRoot) -> bool:
        if not root and not subRoot:
            return True

        if not subRoot:
            return False
        
        def SameTree(n1,n2):
            if not n1 and not n2:
                return True

            if n1 and n2 and n1.val == n2.val:
                return SameTree(n1.left,n2.left) and SameTree(n1.right,n2.right)

            else:

                return False
        
        
        def dfs(root):
            if not root:
                return False

            if root.val == subRoot.val:
                if SameTree(root,subRoot):
                    return True

            if dfs(root.left):
                return True
            if dfs(root.right):
                return True

            return False
            

        return dfs(root)
        
```

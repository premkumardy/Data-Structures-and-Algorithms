338. Counting Bits

Given an integer n, return an array ans of length n + 1 such that for each i (0 <= i <= n), ans[i] is the number of 1's in the binary representation of i.

 

Example 1:

Input: n = 2
Output: [0,1,1]
Explanation:
0 --> 0
1 --> 1
2 --> 10
Example 2:

Input: n = 5
Output: [0,1,1,2,1,2]
Explanation:
0 --> 0
1 --> 1
2 --> 10
3 --> 11
4 --> 100
5 --> 101
 

Constraints:

0 <= n <= 105

```python
#bottom up dp
class Solution(object):
    def countBits(self, n):
        """
        requirement is to return for 0 to n, return a list containing the count of 1's in respective binary value
        refer back to LC 191 number of 1 Bits

        Observation: we know that num x will have last 1 bit more than num y (num y is half of num x, y = (x//2))
        example if we take x as 4 (0100) then num y is 2 (0010)
        here the num 4 is even, so the last bit missing in 2 was 0, both number of 1 bits in x and y matches

        in case of odd number like 5 we know that last bit will be 1 and it will be missing last bit in y
        5 (0101) ,  4 (0100)
        here we can observe the pattern that we can calculate number of 1 by having number of 1's x//2 number 
        and adding 1 to the count if our current number is odd or add 0 or ski if our current number is even

        by using this we can calculate number of 1's in each numer by refering back the ans arr making this run in O(n) TC
        instead of O(n*32) or O(n log n) time complexity

        """
        ans = [0]*(n+1)
        for i in range(n+1):
            ans[i] = (i&1) + ans[i>>1]
        return ans


        
Solution().countBits(15)
```

```python
class Solution(object):
    def countBits(self, n):
        """
        requirement is to return for 0 to n, return a list containing the count of 1's in respective binary value
        refer back to LC 191 number of 1 Bits
        """
        ans = []
        
        for i in range(n+1):
            c = 0
            while i:
                c += i%2
                i = i >> 1
            ans.append(c)
        return ans
Solution().countBits(15)
```

```python
class Solution(object):
    def countBits(self, n):
        """
        :type n: int
        :rtype: List[int]
        """
        ans = []
        
        for i in range(n+1):
            ans.append(bin(i)[2:].count('1'))
        return ans
                
Solution().countBits(15)
```

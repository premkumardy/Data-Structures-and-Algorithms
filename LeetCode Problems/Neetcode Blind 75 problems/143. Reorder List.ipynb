143. Reorder List

You are given the head of a singly linked-list. The list can be represented as:

L0  L1    Ln - 1  Ln
Reorder the list to be on the following form:

L0  Ln  L1  Ln - 1  L2  Ln - 2  
You may not modify the values in the list's nodes. Only nodes themselves may be changed.

 

Example 1:



Input: head = [1,2,3,4]
Output: [1,4,2,3]
Example 2:



Input: head = [1,2,3,4,5]
Output: [1,5,2,4,3]
 

Constraints:

The number of nodes in the list is in the range [1, 5 * 104].
1 <= Node.val <= 1000

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reorderList(self, head):
        """
        Do not return anything, modify head in-place instead.
        """
        slow = head
        fast = head.next
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        mid = slow.next
        slow.next = None
        
        #reverse the second half
        prev = None
        while mid:
            tmp = mid.next
            mid.next = prev
            prev = mid
            mid = tmp
        
        curr = head
        while prev:
            tmp1 = curr.next
            tmp2 = prev.next

            curr.next = prev
            prev.next = tmp1

            curr = tmp1
            prev = tmp2
            
        
        return head
        

        
            
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
head = ListNode()
curr = head
for i in [1,2,3,4]:

    curr.next = ListNode(i)
    curr = curr.next



root = Solution().reorderList(head.next)

vals = []
while root:
    vals.append(str(root.val))
    root = root.next
print("->".join(vals))
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reorderList(self, head):
        """
        Do not return anything, modify head in-place instead.
        """
        if not head:
            return
        

        nodes = []

        curr = head
        while curr:
            nodes.append(curr)
            curr = curr.next
        head1 = ListNode()
        curr = head1
        for i in range(len(nodes)//2):
            j = len(nodes) - 1 -i
            curr.next = nodes[i]
            nodes[i].next = nodes[j]
            curr = nodes[j]
        
        if len(nodes) & 1:
            curr.next = nodes[i+1]
            curr = curr.next
        curr.next = None

        return head1.next
            
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
head = ListNode()
curr = head
for i in range(1,6):

    curr.next = ListNode(i)
    curr = curr.next



root = Solution().reorderList(head.next)

vals = []
while root:
    vals.append(str(root.val))
    root = root.next
print("->".join(vals))
```

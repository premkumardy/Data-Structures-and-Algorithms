143. Reorder List

You are given the head of a singly linked-list. The list can be represented as:

L0  L1    Ln - 1  Ln
Reorder the list to be on the following form:

L0  Ln  L1  Ln - 1  L2  Ln - 2  
You may not modify the values in the list's nodes. Only nodes themselves may be changed.

 

Example 1:



Input: head = [1,2,3,4]
Output: [1,4,2,3]
Example 2:



Input: head = [1,2,3,4,5]
Output: [1,5,2,4,3]
 

Constraints:

The number of nodes in the list is in the range [1, 5 * 104].
1 <= Node.val <= 1000

Requirement:-
return head of the node when ith index is followed by n-ith node in the linked list

Intuition:-

We can store the address of nodes in a list and acheive the same result but can be done with O(1) space complexity with fast and slow pointer approach

The core idea is to use fast and slow pointer to find the mid point of the linked list and reverse the second segment and store the begning node of second segment
and follow the same method as merging two sorted list and we know for a fat that second segment will be of same length if its a even length list or len(first segment) - 1 if its a odd length




```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def reorderList(self, head):
        """
        Do not return anything, modify head in-place instead.
        """
        
        if not head:
            return
        
        #finding the mid point
        slow = head
        fast = head.next
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        midpoint = slow.next
        slow.next = None

        #reversing the second half
        prev = None
        while midpoint:
            tmp = midpoint.next
            midpoint.next = prev
            prev = midpoint
            midpoint = tmp

        #inserting second segment into first as first0->second0->fist1->second1
        curr = head
        while prev:
            tmp1 = curr.next
            tmp2 = prev.next
            curr.next = prev
            prev.next = tmp1
            prev = tmp2
            curr = tmp1

        return head

        
            
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
head = ListNode()
curr = head
for i in range(0,6):

    curr.next = ListNode(i)
    curr = curr.next



root = Solution().reorderList(head.next)

vals = []
while root:
    vals.append(str(root.val))
    root = root.next
print("->".join(vals))
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reorderList(self, head):
        """
        Do not return anything, modify head in-place instead.
        """
        if not head:
            return
        

        nodes = []

        curr = head
        while curr:
            nodes.append(curr)
            curr = curr.next
        head1 = ListNode()
        curr = head1
        for i in range(len(nodes)//2):
            j = len(nodes) - 1 -i
            curr.next = nodes[i]
            nodes[i].next = nodes[j]
            curr = nodes[j]
        
        if len(nodes) & 1:
            curr.next = nodes[i+1]
            curr = curr.next
        curr.next = None

        return head1.next
            
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
head = ListNode()
curr = head
for i in range(1,6):

    curr.next = ListNode(i)
    curr = curr.next



root = Solution().reorderList(head.next)

vals = []
while root:
    vals.append(str(root.val))
    root = root.next
print("->".join(vals))
```

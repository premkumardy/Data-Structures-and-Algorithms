1523. Count Odd Numbers in an Interval Range

Given two non-negative integers low and high. Return the count of odd numbers between low and high (inclusive).

 

Example 1:

Input: low = 3, high = 7
Output: 3
Explanation: The odd numbers between 3 and 7 are [3,5,7].
Example 2:

Input: low = 8, high = 10
Output: 1
Explanation: The odd numbers between 8 and 10 are [9].
 

Constraints:

0 <= low <= high <= 10^9

```python
#correct but TC is O(N) needs O(1) TC

class Solution(object):
    def countOdds(self, low, high):
        """
        :type low: int
        :type high: int
        :rtype: int
        """
        count = 0
        for n in range(low,high+1):
            if n & 1:
                count+=1
        return count
```

Intuition for counting odds in a range [low, high]:

Total numbers in the range = n = high - low + 1.

In any n consecutive numbers:

If n is even, exactly half (n//2) are odd.

If n is odd, then:

If the first number (low) is odd, you get one extra odd, so (n//2) + 1.

If low is even, the extra number is even, so just n//2

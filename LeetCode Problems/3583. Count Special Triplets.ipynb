3583. Count Special Triplets


You are given an integer array nums.

A special triplet is defined as a triplet of indices (i, j, k) such that:

0 <= i < j < k < n, where n = nums.length
nums[i] == nums[j] * 2
nums[k] == nums[j] * 2
Return the total number of special triplets in the array.

Since the answer may be large, return it modulo 109 + 7.

 

Example 1:

Input: nums = [6,3,6]

Output: 1

Explanation:

The only special triplet is (i, j, k) = (0, 1, 2), where:

nums[0] = 6, nums[1] = 3, nums[2] = 6
nums[0] = nums[1] * 2 = 3 * 2 = 6
nums[2] = nums[1] * 2 = 3 * 2 = 6
Example 2:

Input: nums = [0,1,0,0]

Output: 1

Explanation:

The only special triplet is (i, j, k) = (0, 2, 3), where:

nums[0] = 0, nums[2] = 0, nums[3] = 0
nums[0] = nums[2] * 2 = 0 * 2 = 0
nums[3] = nums[2] * 2 = 0 * 2 = 0
Example 3:

Input: nums = [8,4,2,8,4]

Output: 2

Explanation:

There are exactly two special triplets:

(i, j, k) = (0, 1, 3)
nums[0] = 8, nums[1] = 4, nums[3] = 8
nums[0] = nums[1] * 2 = 4 * 2 = 8
nums[3] = nums[1] * 2 = 4 * 2 = 8
(i, j, k) = (1, 2, 4)
nums[1] = 4, nums[2] = 2, nums[4] = 4
nums[1] = nums[2] * 2 = 2 * 2 = 4
nums[4] = nums[2] * 2 = 2 * 2 = 4
 

Constraints:

3 <= n == nums.length <= 105
0 <= nums[i] <= 105

```python
#time complexity O(N^3) - not accepted in LC due to input is at scale of 10^5
#making the time complexity 10^15
import math
class Solution(object):
    def specialTriplets(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        ans = 0
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                for k in range(j+1,len(nums)):
                    print(nums[i],nums[j],nums[k])
                    if nums[i] == (nums[j]*2) and nums[k] == (nums[j]*2):
                        ans += 1
        return ans



nums = [0,1,0,0]
Solution().specialTriplets(nums)
```

```python
"""
first all all the nums and its count to a dict called right
left is initially empty

all we want to check is if an indx has nums[i] * 2 to its left and nums[i] * 2 to its right,
then its a special triplet and increment its answer by 1 
trick is if you if count cccurance in right dic in TC O(N) complexity and while iterating through
 n we can check if n is in right, then decremnt by 1, this kida equals to that we can know if current n 's n*2 exist in left and right side of it
at end of each iterating need to add the current n to left, because lefft was initially empty  
"""
#time complexity O(N)
class Solution(object):
    def specialTriplets(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        left,right,ans = {}, {}, 0
        mod = 10**9 + 7
        for n in nums:
            if not n in right: right[n] = 1
            else: right[n] += 1
        
        for n in nums:
            if n in right:
                right[n] -= 1
            if not right[n] > 0:
                del right[n]

            mid = n*2
            if mid in left and mid in right:
                total_occurance = (left[mid]*right[mid]) % mod
                ans = (ans + total_occurance) % mod
            
            if not n in left:
                left[n] = 1
            else:
                left[n] += 1
            
        return ans

    
nums = [28,52,14,28,34,26,14,52]
Solution().specialTriplets(nums)

```

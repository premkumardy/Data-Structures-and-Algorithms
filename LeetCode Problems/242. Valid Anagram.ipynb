
242. Valid Anagram

Given two strings s and t, return true if t is an anagram of s, and false otherwise.

 

Example 1:

Input: s = "anagram", t = "nagaram"

Output: true

Example 2:

Input: s = "rat", t = "car"

Output: false

 

Constraints:

1 <= s.length, t.length <= 5 * 104
s and t consist of lowercase English letters.

```python
#more efficient
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        if len(s) != len(t):
            return False

        for ch in set(s):
            if s.count(ch) != t.count(ch):
                return False
            
        return True
        
Solution().isAnagram("car",'cat')
```

```python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        return sorted(s) == sorted(t)
        
Solution().isAnagram("tac",'cat')
```

```python
class Solution(object):
    def isAnagram(self, s, t):
        if len(s) != len(t):
            return False

        count = {}

        # Fill the hash map
        for char in s:
            count[char] = count.get(char, 0) + 1
        
        # Subtract from the hash map
        for char in t:
            if char not in count or count[char] == 0:
                return False
            count[char] -= 1
            
        return True
        
Solution().isAnagram("tac",'cat')
```

```python
from collections import Counter
class Solution(object):
    def isAnagram(self, s, t):
        return Counter(s) == Counter(t)
Solution().isAnagram("tac",'cat')
```

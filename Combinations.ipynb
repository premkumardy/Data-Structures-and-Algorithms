Combinations

You are given two integers n and k, return all possible combinations of k numbers chosen from the range [1, n].

You may return the answer in any order.

Example 1:

Input: n = 3, k = 2

Output: [
    [1,2],
    [1,3],
    [2,3]
]
Example 2:

Input: n = 3, k = 3

Output: [[1,2,3]]

```python
class Solution:
    def combine(self, n: int, k: int):

        ans,tmp = [],[]

        def dfs(i):

            if len(tmp) == k:
                ans.append(tmp[:])
                return

            if i > n:
                return
            
            tmp.append(i)
            dfs(i+1)
            tmp.pop()
            dfs(i+1)

        dfs(1)
        return ans
Solution().combine(5,2)
```

```python
class Solution:
    def combine(self, n: int, k: int):

        ans = []
        def dfs(i,tmp):
            if len(tmp) == k:
                ans.append(tmp[:])
                return
            if i > n:
                return
            
            for j in range(i,n+1):
                tmp.append(j)
                dfs(j,tmp)
                tmp.pop()
        dfs(1,[])
        return ans
Solution().combine(5,2)
```

Combination Sum
You are given an array of distinct integers nums and a target integer target. Your task is to return a list of all unique combinations of nums where the chosen numbers sum to target.

The same number may be chosen from nums an unlimited number of times. Two combinations are the same if the frequency of each of the chosen numbers is the same, otherwise they are different.

You may return the combinations in any order and the order of the numbers in each combination can be in any order.

Example 1:

Input: 
nums = [2,5,6,9] 
target = 9

Output: [[2,2,5],[9]]
Explanation:
2 + 2 + 5 = 9. We use 2 twice, and 5 once.
9 = 9. We use 9 once.

Example 2:

Input: 
nums = [3,4,5]
target = 16

Output: [[3,3,3,3,4],[3,3,5,5],[4,4,4,4],[3,4,4,5]]
Example 3:

Input: 
nums = [3]
target = 5

Output: []
Constraints:

All elements of nums are distinct.
1 <= nums.length <= 20
2 <= nums[i] <= 30
2 <= target <= 30

```python
class Solution:
    def combinationSum(self, nums, target: int):
        ans = []

        def dfs(i,tmp,total):
            if total == target:
                ans.append(tmp[:])
                return
            if total > target or i >= len(nums):
                return
            tmp.append(nums[i])
            dfs(i,tmp,total + nums[i])

            tmp.pop()
            dfs(i+1,tmp,total)
            
        dfs(0,[],0)
        return ans
    
        
Solution().combinationSum([2,5,6,9],9)
```

```python
class Solution(object):
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """
        ans = []

        def dfs(i,tmp,total):
            if total == target:
                ans.append(tmp[:])
                return
            if i >= len(candidates) or total > target:
                return
            
            for j in range(i,len(candidates)):
                tmp.append(candidates[j])
                dfs(j,tmp,total+candidates[j])
                tmp.pop()
            
        dfs(0,[],0)
        return ans
        
Solution().combinationSum([2,5,6,9],9)
```

Letter Combinations of a Phone Number

Each digit (not including 1) is mapped to a set of characters as shown below:

A digit could represent any one of the characters it maps to.

Return all possible letter combinations that digits could represent. You may return the answer in any order.



Example 1:

Input: digits = "34"

Output: ["dg","dh","di","eg","eh","ei","fg","fh","fi"]
Example 2:

Input: digits = ""

Output: []
Constraints:

0 <= digits.length <= 4
2 <= digits[i] <= 9

```python
class Solution:
    def letterCombinations(self, digits: str):
        res = []
        digitToChar = {
            "2": "abc",
            "3": "def",
            "4": "ghi",
            "5": "jkl",
            "6": "mno",
            "7": "qprs",
            "8": "tuv",
            "9": "wxyz",
        }

        def backtrack(i, curStr):
            if len(curStr) == len(digits):
                res.append(curStr)
                return
            for c in digitToChar[digits[i]]:
                backtrack(i + 1, curStr + c)

        if digits:
            backtrack(0, "")

        return res
    
Solution().letterCombinations('254')
```

```python
class Solution:
    def letterCombinations(self, digits):
        if not digits:
            return []

        res = [""]
        digitToChar = {
            "2": "abc",
            "3": "def",
            "4": "ghi",
            "5": "jkl",
            "6": "mno",
            "7": "qprs",
            "8": "tuv",
            "9": "wxyz",
        }

        for digit in digits:
            tmp = []
            for curStr in res:
                for c in digitToChar[digit]:
                    tmp.append(curStr + c)
            res = tmp
        return res
Solution().letterCombinations('23')
```
